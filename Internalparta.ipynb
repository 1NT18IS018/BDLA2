```python
#linear regression
import pandas as pd
import scipy.stats as sp
import statistics as st
import matplotlib.pyplot as plt
names=['x','y']
data=pd.read_csv("lin.csv",names=names)
x=data['x']
y=data['y']
mx=st.mean(x)
my=st.mean(y)
sx=st.stdev(x)
sy=st.stdev(y)
r=sp.pearsonr(x,y)[0]
m=r*sy/sx
c=my-m*mx
y_pred=m*x+c
print(y_pred,y)
plt.scatter(x,y,color="red")
plt.plot(x,y_pred,color="cyan")
plt.show
#sse
se=(y-y_pred)**2
sse=sum(se)
print(sse)
#ssr
sr=(y_pred-my)**2
ssr=sum(sr)
print(ssr)
#sst
st=(y-my)**2
sst=sum(st)
print(sst)
#r
r=ssr/sst
print(r)
```

```python
#prg1
import numpy as np
import statistics as st
import math
arr=list(map(int,input().split()))
#mean
mean=sum(arr)/len(arr)
print(mean)
print(st.mean(arr))
```

```python
#median
arr.sort()
print(arr)
if(len(arr)%2==0):
    med=(arr[(len(arr)//2)-1]+arr[len(arr)//2])/2
else:
    med=arr[(len(arr)//2)]
print(med)
print(st.median(arr))
#mode
i=0
l=[]
c=[]
while(i<len(arr)):
    count=0
    j=0
    while(j<len(arr)):
        if(arr[j]==arr[i]):
            count+=1
        j+=1
    if(count>1):
        c.append(count)
        l.append(arr[i])
    i+=1
ind=c.index(max(c))
mode=l[ind]
print(mode)
print(st.mode(arr))
#stdev and variation
sums=0
for i in arr:
    sums+=(i-mean)**2
v=sums/(len(arr)-1)
print(v)
print(st.variance(arr))
print(math.sqrt(v))
print(st.stdev(arr))
#normalization
mi=min(arr)
ma=max(arr)
narr=[]
for i in arr:
    n=(i-mi)/(ma-mi)
    narr.append(n)
print(narr)
print(st.mean(narr))
print(st.stdev(narr))
#standardization
sarr=[]
for i in arr:
    n=(i-mean)/(st.stdev(arr))
    sarr.append(n)
print(sarr)
print(st.mean(sarr))
print(st.stdev(sarr))
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import StandardScaler
data=[[81,52,33],[43,51,26],[21,32,54]]
n=Normalizer().fit(data)
n=n.transform(data)
print(n)
s=StandardScaler().fit(data)
s=s.transform(data)
print(s)
print(np.mean(s))
print(np.std(s))
```

```python
#Kmeans
import numpy as np
from copy import deepcopy
import pandas as pd
def euclidean(a,b,ax=1):
    return np.linalg.norm(a-b,axis=ax)
data=pd.read_csv("kmean.csv")
x=data['x']
y=data['y']
X=np.array(list(zip(x,y)))
C_x = [6.2, 6.6 ,6.5]
C_y = [3.2, 3.7, 3.0]
centroid=np.array(list(zip(C_x,C_y)))
old_centroid=np.zeros(centroid.shape)
clusters=np.zeros(len(X))
error=euclidean(centroid,old_centroid,None)
print(error)
itter=0
k=3
while(error!=0):
    itter+=1
    for i in range(len(X)):
        distances=euclidean(X[i],centroid)
        cluster=np.argmin(distances)
        clusters[i]=cluster
    old_centroid=deepcopy(centroid)
    for p in range(k):
        points=[X[j] for j in range(len(X)) if clusters[j]==p]
        centroid[p]=np.mean(points,axis=0)
    error=euclidean(centroid,old_centroid,None)
    print("iteration ",itter,"centroid :")
    print(centroid)
    print(error)
```
